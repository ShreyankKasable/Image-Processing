### HOW TO ADD A LOGO ON A LIVE VIDEO USING OPENCV

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
logo = cv.imread("LOGO.jpg")
logo = cv.resize(logo, (int(360/2), int(360/2)))

cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    img = cv.resize(img, (int(1280),int(720)))
    
    img[0:180,1100:1280] = logo

    cv.imshow("IMG", img)
    
    key = cv.waitKey(20)
    
    if key == 27:
        break
cam.release()
cv.destroyAllWindows()
```

### HOW TO AUTO FIT LOGO IN OPENCV

```python
logo = cv.imread("download.png")
logo = cv.resize(logo, (50, 50))
cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    img = cv.resize(img, (int(1280),int(720)))
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    margin_top  = int(ht/100)             # 1%
    logo_height = int((ht/100)*7)         # 7%
    
    margin_left = int((wt/100)*95.5)      # 95.5%
    logo_width  = int((wt/100)*4)         # 4%
    
    logo = cv.resize(logo, (logo_width, logo_width))
    
    
    img[ margin_top : logo_width + margin_top , margin_left : margin_left + logo_width ] = logo

    cv.imshow("IMG", img)
    
    key = cv.waitKey(20)
    
    if key == 27:
        break
cam.release()
cv.destroyAllWindows()
```

### HOW TO FIT A LOGO IN OPENCV

```python
logo = cv.imread("download.png")
logo = cv.resize(logo, (50, 50))
cam = cv.VideoCapture(0)
scaling_factor = 1.0

while True:
    logo = cv.imread("download.png")
    _, img = cam.read()
    img = cv.flip(img, 1)
    
    img = cv.resize(img, (int(img.shape[1]*scaling_factor), int(img.shape[0]*scaling_factor)))
    ht = img.shape[0]
    wt = img.shape[1]
    
    margin_top  = int(ht/100)             # 1%
    logo_height = int((ht/100)*7)         # 7%
    margin_left = int((wt/100)*95.5)      # 95.5%
    logo_width  = int((wt/100)*4)         # 4%
    
    side = min(logo_width, logo_height)
    logo = cv.resize(logo, (logo_width, logo_width))
    img[ margin_top : logo_width + margin_top , margin_left : margin_left + logo_width ] = logo

    cv.imshow("IMG", img)
    
    key = cv.waitKey(20)
    
    if key == 27:
        break
    elif key == 115:
        if scaling_factor>0:
            scaling_factor *= 0.9
    elif key == 119:
        if scaling_factor<720:
            scaling_factor *= 1.1   
            
cam.release()
cv.destroyAllWindows()
```

### HOW TO ADD TRANSPERENCY TO A LOGO IN OPENCV

```python
logo = cv.imread("download.png")
logo = cv.resize(logo, (50, 50))
cam = cv.VideoCapture(0)

while True:
    logo = cv.imread("download.png")
    _, img = cam.read()
    img = cv.flip(img, 1)
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    margin_top  = int(ht/100)             # 1%
    logo_height = int((ht/100)*7)         # 7%
    margin_left = int((wt/100)*95.5)      # 95.5%
    logo_width  = int((wt/100)*4)         # 4%
    
    side = min(logo_width, logo_height)
    
    bg = img[margin_top : side + margin_top , margin_left : margin_left + side].copy()
    logo = cv.resize(logo, (logo_width, logo_width))
    
    final_logo = cv.addWeighted(bg, .5, logo, .5,0)
    img[ margin_top : logo_width + margin_top , margin_left : margin_left + logo_width ] = final_logo

    cv.imshow("IMG", img)
    #cv.imshow("BG", bg)
    
    key = cv.waitKey(20)
    
    if key == 27:
        break 
            
cam.release()
cv.destroyAllWindows()
```
