```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

### LOADING AN IMAGE

```python
img = cv.imread('RGB.jpg')

img = cv.cvtColor(img, cv.COLOR_RGB2BGR)

plt.imshow(img)
```

```python
plt.imshow(img[:1000 , : , :])
```

```python
plt.imshow(img[500:1000 , : , :])
```

```python
plt.imshow(img[1000: , : , :])
```

```python
plt.imshow(img[: , : , 1])
```

```python
image = plt.imshow(img[: , : , 0])

cv.imwrite("image.jpg", img[:,:,0])
```

```python
plt.imshow(img[: , : , 2])
```
